```python
import math

# Lista de meses dos 6 primeiros meses do ano
meses = ['Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

# Crie uma figura com 1 linha e 6 colunas (um para cada um dos 6 primeiros meses)
fig, axs = plt.subplots(1, 6, figsize=(15, 5))  # Ajuste o tamanho da figura conforme necessário

# Para cada mês, calcule o número de classes usando a regra de Sturges
# e plote o histograma na subparcela correspondente
for i, temperatura in enumerate([
    temperatura_Julho, temperatura_Agosto, temperatura_Setembro, temperatura_Outubro, temperatura_Novembro, temperatura_Dezembro
]):
    n = len(temperatura)  # Número de observações para o mês atual
    k = 1 + math.log2(n)  # Usar a regra de Sturges para calcular o número de classes
    k = int(round(k))  # Arredonde para o número inteiro mais próximo

    axs[i].hist(temperatura, bins=k, edgecolor='black')
    axs[i].set_xlabel('Temperatura')
    axs[i].set_ylabel('Contagem')
    axs[i].set_title(f'Histograma de {meses[i]}')

# Ajuste o espaçamento entre as subparcelas para evitar sobreposição de rótulos
plt.tight_layout()

# Exiba a figura
plt.show()
```

```python
import numpy as np
from scipy.stats import weibull_min, kstest, cramervonmises, anderson, ttest_1samp
from tabulate import tabulate

def weibull_moments(shape, scale, moment):
    return scale ** moment * np.math.gamma(1 + moment / shape)

# Suponha que você tem listas de dados de temperatura para cada mês
temperatura_meses = [
    temperatura_Janeiro, temperatura_Fevereiro, temperatura_Março,
    temperatura_Abril, temperatura_Maio, temperatura_Junho,
    temperatura_Julho, temperatura_Agosto, temperatura_Setembro,
    temperatura_Outubro, temperatura_Novembro, temperatura_Dezembro
]

# Criação de listas para armazenar os resultados
meses = [f'Mês {i+1}' for i in range(12)]
testes = ['KS', 'CvM', 'AD', 't-Student']
resultados = []

# Valor de referência para o teste t-Student
valor_referencia = 0  # Substitua pelo valor desejado

# Nível de significância para o teste t-Student
significancia = 0.05

# Loop para realizar os testes para cada mês e cada distribuição
for i, temperatura in enumerate(temperatura_meses):
    resultados_mes = []
    interpretacao = "NS"  # Defina um valor padrão

    for teste in testes:
        if teste == 'KS':
            ks_stat, ks_p_value = kstest(temperatura, 'weibull_min', args=weibull_min.fit(temperatura))
            resultados_mes.extend([ks_stat, ks_p_value])
        elif teste == 'CvM':
            cvm_result = cramervonmises(temperatura, 'weibull_min', args=weibull_min.fit(temperatura))
            cvm_stat = cvm_result.statistic
            cvm_p_value = cvm_result.pvalue
            resultados_mes.extend([cvm_stat, cvm_p_value])
        elif teste == 'AD':
            ad_statistic, ad_critical_values, ad_significance_level = anderson(temperatura, dist='weibull_min')
            interpretacao = "NS"  # NS por padrão (não segue)
            for j, critical_value in enumerate(ad_critical_values):
                if ad_statistic < critical_value:
                    interpretacao = f'{ad_significance_level[j]:.2f}'
                    break
            resultados_mes.extend([ad_statistic, interpretacao])
        elif teste == 't-Student':
            # Teste t-Student
            t_stat, t_p_value = ttest_1samp(temperatura, valor_referencia)
            resultado_t_student = "Segue t-Student" if t_p_value > significancia else "Não Segue t-Student"
            resultados_mes.extend([t_stat, t_p_value, resultado_t_student])

    # Adiciona a coluna indicando se segue ou não a distribuição Weibull
    resultados_mes.append("Segue" if interpretacao != "NS" else "Não Segue")

    # Adiciona a coluna da FGM apenas para os meses que seguem a distribuição Weibull
    if interpretacao != "NS":
        shape, loc, scale = weibull_min.fit(temperatura)
        fgm = weibull_moments(shape, scale, moment=1)
        resultados_mes.append(fgm)
    else:
        resultados_mes.append("Não Calculado")

    resultados.append(resultados_mes)

# Conversão de listas para arrays NumPy com dtype de objeto
meses = np.array(meses, dtype=object)
resultados = np.array(resultados, dtype=object)

# Criação da tabela
header = ["Mês", "KS Estatística", "KS Valor-p", "CvM Estatística", "CvM Valor-p", "AD Estatística", "AD Significância", "t-Student Estatística", "t-Student Valor-p", "t-Student Resultado",  "FGM", "Segue Weibull?"]
tabela_resultados = np.column_stack([meses, resultados])

# Formatação dos dados para a tabela
tabela_formatada = []
for linha in tabela_resultados:
    linha_formatada = [linha[0]] + [f'{elem:.4f}' if isinstance(elem, (float, np.float_)) else elem for elem in linha[1:]]
    tabela_formatada.append(linha_formatada)

# Exibição da tabela
print(tabulate(tabela_formatada, headers=header, tablefmt="pretty"))
```

```python
import numpy as np
from scipy.stats import weibull_min, kstest, ttest_1samp
from tabulate import tabulate

def weibull_moments(shape, scale, moment):
    return scale ** moment * np.math.gamma(1 + moment / shape)

# Suponha que você tem listas de dados de temperatura para cada mês
temperatura_meses = [
    temperatura_Janeiro, temperatura_Fevereiro, temperatura_Março,
    temperatura_Abril, temperatura_Maio, temperatura_Junho,
    temperatura_Julho, temperatura_Agosto, temperatura_Setembro,
    temperatura_Outubro, temperatura_Novembro, temperatura_Dezembro
]

# Criação de listas para armazenar os resultados
meses = [f'Mês {i+1}' for i in range(12)]
resultados = []

# Valor de referência para o teste t-Student
valor_referencia = 0  # Substitua pelo valor desejado

# Nível de significância para o teste t-Student
significancia = 0.05

# Loop para realizar os testes para cada mês e cada distribuição
for i, temperatura in enumerate(temperatura_meses):
    resultados_mes = []

    # Teste de Kolmogorov-Smirnov (KS) para Weibull
    ks_stat, ks_p_value = kstest(temperatura, 'weibull_min', args=weibull_min.fit(temperatura))
    resultados_mes.extend([ks_stat, ks_p_value, "Segue Weibull" if ks_p_value > significancia else "Não Segue Weibull"])

    # Teste t-Student
    t_stat, t_p_value = ttest_1samp(temperatura, valor_referencia)
    resultados_mes.extend(["Segue t-Student" if t_p_value > significancia else "Não Segue t-Student"])

    # Adiciona a coluna da FGM para os meses que seguem a distribuição Weibull
    if ks_p_value > significancia:
        shape, loc, scale = weibull_min.fit(temperatura)
        fgm1 = weibull_moments(shape, scale, moment=1)
        fgm2 = weibull_moments(shape, scale, moment=2)
        fgm3 = weibull_moments(shape, scale, moment=3)
        fgm4 = weibull_moments(shape, scale, moment=4)
        resultados_mes.extend([fgm1, fgm2, fgm3, fgm4])
    else:
        resultados_mes.extend(["Não Calculado"] * 4)

    resultados.append(resultados_mes)

# Conversão de listas para arrays NumPy com dtype de objeto
meses = np.array(meses, dtype=object)
resultados = np.array(resultados, dtype=object)

# Criação da tabela
header = ["Mês", "KS Estatística", "KS Valor-p", "Segue Weibull?", "t-Student Resultado", "FGM 1", "FGM 2", "FGM 3", "FGM 4"]
tabela_resultados = np.column_stack([meses, resultados])

# Formatação dos dados para a tabela
tabela_formatada = []
for linha in tabela_resultados:
    linha_formatada = [linha[0]] + [f'{elem:.4f}' if isinstance(elem, (float, np.float_)) else elem for elem in linha[1:]]
    tabela_formatada.append(linha_formatada)

# Exibição da tabela
print(tabulate(tabela_formatada, headers=header, tablefmt="pretty"))
```

Mês: Indica o número do mês.

**KS Estatística** (Kolmogorov-Smirnov): Representa a estatística KS, que é usada para testar a hipótese nula de que duas distribuições contínuas são idênticas. É uma medida da maior discrepância entre as funções de distribuição acumulada das duas amostras.

**KS Valor-p**: É o valor-p associado ao teste de Kolmogorov-Smirnov. Indica a probabilidade de observar uma estatística KS tão extrema quanto a calculada, sob a hipótese nula de que as amostras são idênticas.

**CvM Estatística** (Cramér-von Mises): Representa a estatística de Cramér-von Mises, usada para testar a hipótese nula de que uma amostra é retirada de uma população com uma distribuição de probabilidade específica.

**CvM Valor-p**: É o valor-p associado ao teste de Cramér-von Mises. Indica a probabilidade de observar uma estatística de CvM tão extrema quanto a calculada, sob a hipótese nula.

**AD Estatística** (Anderson-Darling): Representa a estatística de Anderson-Darling, um teste de bondade de ajuste que é uma variação do teste de Kolmogorov-Smirnov, dando mais peso às caudas da distribuição.

**AD Significância:** Indica se a estatística de Anderson-Darling é estatisticamente significativa. "NS" significa não significativo.

**t-Student Estatística**: Refere-se à estatística do teste t de Student. É utilizado para testar se há diferença significativa entre as médias de duas amostras independentes.

**t-Student Valor-p**: É o valor-p associado ao teste t de Student. Indica a probabilidade de observar uma estatística t tão extrema quanto a calculada, sob a hipótese nula de que as médias das amostras são iguais.

**t-Student Resultado**: Indica se o teste t de Student foi considerado significativo.

**FGM:** Indica se a distribuição segue uma distribuição específica chamada Função geratriz de momento.

**Segue Weibull?**: Indica se a distribuição segue a distribuição de Weibull. "Segue" significa que a distribuição segue Weibull, enquanto "Não Segue" indica o contrário. "Não Calculado" pode indicar que o teste específico para Weibull não foi realizado.

```python
import numpy as np
from scipy.stats import weibull_min, kstest, cramervonmises, anderson, ttest_1samp
from tabulate import tabulate

def weibull_moments(shape, scale, moment):
    return scale ** moment * np.math.gamma(1 + moment / shape)

# Suponha que você tem listas de dados de temperatura para cada mês
temperatura_meses = [
    temperatura_Janeiro, temperatura_Fevereiro, temperatura_Março,
    temperatura_Abril, temperatura_Maio, temperatura_Junho,
    temperatura_Julho, temperatura_Agosto, temperatura_Setembro,
    temperatura_Outubro, temperatura_Novembro, temperatura_Dezembro
]

# Criação de listas para armazenar os resultados
meses = [f'Mês {i+1}' for i in range(12)]
testes = ['KS', 'CvM', 'AD', 't-Student']
resultados = []

# Valor de referência para o teste t-Student
valor_referencia = 0  # Substitua pelo valor desejado

# Nível de significância para o teste t-Student
significancia = 0.05

# Loop para realizar os testes para cada mês e cada distribuição
for i, temperatura in enumerate(temperatura_meses):
    resultados_mes = []
    for teste in testes:
        if teste == 'KS':
            ks_stat, ks_p_value = kstest(temperatura, 'weibull_min', args=weibull_min.fit(temperatura))
            resultados_mes.extend([ks_stat, ks_p_value])
        elif teste == 'CvM':
            cvm_result = cramervonmises(temperatura, 'weibull_min', args=weibull_min.fit(temperatura))
            cvm_stat = cvm_result.statistic
            cvm_p_value = cvm_result.pvalue
            resultados_mes.extend([cvm_stat, cvm_p_value])
        elif teste == 'AD':
            ad_statistic, ad_critical_values, ad_significance_level = anderson(temperatura, dist='weibull_min')
            interpretacao = "NS"  # NS por padrão (não segue)
            for j, critical_value in enumerate(ad_critical_values):
                if ad_statistic < critical_value:
                    interpretacao = f'{ad_significance_level[j]:.2f}'
                    break
            resultados_mes.extend([ad_statistic, interpretacao])
        elif teste == 't-Student':
            t_stat, t_p_value = ttest_1samp(temperatura, valor_referencia)
            resultados_mes.extend([t_stat, t_p_value, "Significativo" if t_p_value < significancia else "Não Significativo"])

    # Adiciona a coluna indicando se segue ou não a distribuição Weibull
    resultados_mes.append("Segue" if interpretacao != "NS" else "Não Segue")

    # Adiciona a coluna da FGM apenas para os meses que seguem a distribuição Weibull
    if interpretacao != "NS":
        shape, loc, scale = weibull_min.fit(temperatura)
        fgm = weibull_moments(shape, scale, moment=1)
        resultados_mes.append(fgm)
    else:
        resultados_mes.append("Não Calculado")

    resultados.append(resultados_mes)

# Conversão de listas para arrays NumPy com dtype de objeto
meses = np.array(meses, dtype=object)
resultados = np.array(resultados, dtype=object)

header = ["Mês", "KS Estatística", "KS Valor-p", "CvM Estatística", "CvM Valor-p", "AD Estatística", "AD Significância"]
tabela_resultados = np.column_stack([meses, resultados[:, :6]])

# Formatação dos dados para a tabela
tabela_formatada = []
for linha in tabela_resultados:
    linha_formatada = [linha[0]] + [f'{elem:.4f}' if isinstance(elem, (float, np.float_)) else elem for elem in linha[1:]]
    tabela_formatada.append(linha_formatada)

# Exibição da tabela
print(tabulate(tabela_formatada, headers=header, tablefmt="pretty"))
```
